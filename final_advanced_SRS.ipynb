# Importing libraries and datasets


```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive', force_remount = True)
directory = "/content/drive/MyDrive/Public_folder_Yahoo_RS/"

### Genres ###
genres_columns = ["Id", "Parent", "Level", "Name"]
genres = pd.read_csv(directory + "genre-hierarchy.txt", header = None, sep = '\t', names = genres_columns)
### Songs ###
songs_columns = ["Id", "Album", "Artist", "Genre"]
songs = pd.read_csv(directory + "song-attributes.txt", header = None, sep = '\t', names = songs_columns)
### Ratings ###
ratings_columns = ["User", "Song", "Rating"]
ratings = pd.read_csv(directory + "train_0.txt", header = None, sep = '\t', names = ratings_columns)
print(ratings)
```

# Analyzing the users and finding the ones with the most ratings

```python
users_num_ratings = []
cnt = 0

for i in range(int(len(ratings['User']) / 30)): 
    if(ratings['User'][i] == ratings['User'][i + 1]):
        cnt += 1
    else:
        users_num_ratings.append([ratings['User'][i], cnt + 1])
        cnt = 0

print("Total users analyzed : " + str(len(users_num_ratings)))
users_num_ratings.sort(key = lambda x: x[1], reverse = True)
users_num_ratings = users_num_ratings[0:1000]
users_num_ratings.sort(key = lambda x: x[0], reverse = True)
print(users_num_ratings)
participant_users = []
for user in users_num_ratings:
    participant_users.append(user[0])
print(participant_users)
```

# Filtering the ratings and keeping only the useful ones

```python
import numpy as np

np_ratings = np.array(ratings)

useful_ratings = []
max_limit = max(participant_users)
index = 0
current = participant_users.pop()
while(np_ratings[index][0] < max_limit + 1):
    if(np_ratings[index][0] == current):
        useful_ratings.append(np_ratings[index])
    if(np_ratings[index][0] > current):
        if(participant_users != []):
            current = participant_users.pop()
        else:
            break
    index = index + 1

np_useful_ratings = np.array(useful_ratings)
print(np_useful_ratings)
```

# Verifing the number of users we work with

```python
users_num_ratings = []
cnt = 0

for j in range(len(np_useful_ratings)):
    if(j == len(np_useful_ratings) - 1):
        users_num_ratings.append([np_useful_ratings[j][0], cnt + 1])
    else:
        if(np_useful_ratings[j][0] == np_useful_ratings[j + 1][0]):
            cnt += 1
        else:
            users_num_ratings.append([np_useful_ratings[j][0], cnt + 1])
            cnt = 0

print("Total users analyzed : " + str(len(users_num_ratings)))
print(users_num_ratings)
participant_users = []
for user in users_num_ratings:
    participant_users.append(user[0])
print(participant_users)
```

# Preprocessing the genre and the song table

```python
### Finding the root of each genre ###
np_genres = np.array(genres)
roots = []
for genre in np_genres:
    current = genre[0]
    parent = genre[1]
    while(current != parent):
        current = parent
        parent = np_genres[current][1]
    roots.append(current)
print("Genre with id = 8 : " + str(np_genres[7][3]))
print("The root of this genre is : " + str(np_genres[roots[7]][3]))
```

```python
### Replacing the genre of every song with the root of the genre ###
np_songs = np.array(songs)
for song in range(len(np_songs)):
    np_songs[song][3] = roots[np_songs[song][3]]
```

# Computing the mean(ratings) by user and genre

```python
### These two dictionaries will help us compute the mean(rating) of every user, depending on the genre of the song ###
### ratings_sum_by_genre : {'i-j': a} means that the user i has total sum(rating) = a in genre j ###
### ratings_by_genre : {'i-j': a} means that the user i has rated a songs with genre j ###
ratings_sum_by_genre = {}
ratings_by_genre = {}
for user in participant_users:
    for genre in np.sort(np.unique(np.array(roots))):
        ratings_sum_by_genre[str(user) + "-" + str(genre)] = 0
        ratings_by_genre[str(user) + "-" + str(genre)] = 0

participant_users.reverse()
max_limit = max(participant_users)
index = 0
current = participant_users.pop()
while(np_ratings[index][0] < max_limit + 1):
    if(np_ratings[index][0] == current):
        ratings_sum_by_genre[str(np_ratings[index][0]) + "-" + str(np_songs[np_ratings[index][1]][3])] += np.ushort(np_ratings[index][2])
        ratings_by_genre[str(np_ratings[index][0]) + "-" + str(np_songs[np_ratings[index][1]][3])] += 1
    if(np_ratings[index][0] > current):
        if(participant_users != []):
            current = participant_users.pop()
        else:
            break
    index = index + 1

### Finally the division (ratings_sum_by_genre / ratings_by_genre) will give us the mean(rating) by user and genre ###
for rating in ratings_sum_by_genre:
    if(ratings_by_genre[rating] != 0):
        ratings_sum_by_genre[rating] /= ratings_by_genre[rating]
```

```python
users_dictionary = {}
for j in range(len(participant_users)):
    users_dictionary[j] = participant_users[j]

songs_dictionary = {}
for j in range(len(np_total_songs)):
    songs_dictionary[j] = np_total_songs[j]

for user in range(len(participant_users)):
    for song in range(len(np_total_songs)):
        if(np_matrix[user][song] == 0):
            np_matrix[user][song] = ratings_sum_by_genre[str(users_dictionary[user]) + "-" + str(np_songs[songs_dictionary[song]][3])]

df_matrix = pd.DataFrame(np_matrix, columns = np_total_songs, index = participant_users)
df_matrix
```

# Completing the rest gaps of the matrix with the mean(ratings) by user only

```python
for user in np_matrix:
    m = np.mean(user[user > 0])
    user[user == 0] = m

print(np_matrix)
total_cells = len(participant_users) * len(np_total_songs)
print("Total cells: " + str(total_cells))
non_zero_cells = np.count_nonzero(np_matrix)
print("Total non zeros: " + str(non_zero_cells))
print("Positive percentage: " + str(non_zero_cells / total_cells))
```

# Applying the SVD transformation

```python
KapaParameter = 100   
U, S, VT = np.linalg.svd(np_matrix, full_matrices = False)
U = U[:, :KapaParameter]
S = S[:KapaParameter]
V = VT[:KapaParameter, :].T
```

# Computing similarities between users

```python
from scipy import spatial

total_similarities = []
for user0 in range(len(participant_users)):
    similarities = []
    for user1 in range(len(participant_users)): 
        similarities.append(1 - spatial.distance.cosine(U[user0], U[user1]))
    total_similarities.append(similarities)

np_total_similarities = np.array(total_similarities)
print(np_total_similarities[0][10])
```

# Defining the n most similar users for every user

```python
n_closest = 5
total_similarar_users = []
for user0 in participant_users:
    similarities_list_with_index = []
    for user1 in participant_users:
        similarities_list_with_index.append([user1, df_similarities.loc[user0][user1]])
    similarities_list_with_index.sort(key = lambda x: x[1], reverse = True)
    similar_users = similarities_list_with_index[:n_closest]
    total_similarar_users.append(similar_users)

df_similar_users = pd.DataFrame(total_similarar_users, columns = range(n_closest), index = participant_users)
df_similar_users
```

# Predicting ratings based on the similarities and the matrix

```python
predicted_ratings = []
for rating in np_useful_test_ratings:
    if rating[1] in np_total_songs:
        prediction = 0
        denominator = 0
        for similar_user in df_similar_users.loc[rating[0]]:
            prediction += df_matrix.loc[similar_user[0]][rating[1]] * similar_user[1]
            denominator += similar_user[1]
        predicted_ratings.append([rating[0], rating[1], prediction / denominator, rating[2]])

print(predicted_ratings)
print("Prediction : The user with id = " + str(predicted_ratings[0][0]) + " would rate with " + str(predicted_ratings[0][2]) + " the song with id = " + str(predicted_ratings[0][1] + 1) + ", while the actual rating is " + str(predicted_ratings[0][3]))
```

# Computing the mean(ratings) by song

```python
np_matrix_T = np_matrix.T
means_by_song = []

for j in range(len(np_matrix_T)):
    means_by_song.append(np.mean(np_matrix_T[j]))

means_by_song[0]
```

# Useful dictionaries : We need the transfers ***user_id*** <-> **matrix_row** and **song_id** <-> ***matrix_column***

```python
users_dictionary_reversed = {}
for j in range(len(participant_users)):
    users_dictionary_reversed[participant_users[j]] = j

songs_dictionary_reversed = {}
for j in range(len(np_total_songs)):
    songs_dictionary_reversed[np_total_songs[j]] = j
```

# Giving the 50% of our prediction to the mean(ratings) by song

```python
for rating in predicted_ratings:
    rating[2] = (rating[2] + means_by_song[songs_dictionary_reversed[rating[1]]]) / 2

print(predicted_ratings)
print("Prediction : The user with id = " + str(predicted_ratings[0][0]) + " would rate with " + str(predicted_ratings[0][2]) + " the song with id = " + str(predicted_ratings[0][1] + 1) + ", while the actual rating is " + str(predicted_ratings[0][3]))
```

# Defining the mean squared error of our prediction

```python
import math

total_error = 0
for prediction in predicted_ratings:
    total_error += (prediction[3] - prediction[2]) ** 2
mean_error = total_error / len(predicted_ratings)
print(mean_error)
```
